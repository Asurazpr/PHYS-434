```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy import special
from scipy.stats import norm
```

```python
plt.rcParams["figure.figsize"] = (30,10)
```

```python
# A little statistics
# 1. B.
x = np.linspace(-3, 3)
plt.plot(x, special.erfc(x))
plt.xlabel('$x$')
plt.ylabel('$erfc(x)$')
plt.show()
# the erfc function which represents the integration of the 
# Gaussian distribution.
```

```python
# 1. B. cont.
d = stats.norm.rvs(loc = 0, scale = 2, size = 100000)
fig, ax = plt.subplots(1, 1)
plt.tick_params(labelsize = 24)
plt.xlim([-1,1])
x = np.linspace(-1,1,200)
ax.plot(x,stats.norm.cdf(x,loc = 0, scale = 2),linewidth = 8,alpha = 0.7)
plt.xlabel('$x$')
plt.ylabel('$cdf(x)$')
plt.show()
# The plot of Gaussian distribution's cdf with mean at 0 and variance at 2.
```

```python
# 1. B. cont.
y1 = stats.norm.cdf(x,loc = 0, scale = 0.25)
y1
# This data with mean at 0 and 
# variance equal to 0.25 roughly matches the z-table data.
```

```python
# 1. C.
sigma = stats.norm.ppf(1 - y1,loc = 0, scale = 0.25)
sigma
# In the standard normal distribution, the 
# current distribution has its right half's probabilities
# is smaller than the mean which is accurate 
# considering we are using the x values from -1 to 1,
# and we are substracting it by 1 to acquire the sigma.
```

```python
# 2. B. 
# This is a distribution from 0.01 to 0.99 of
# Lognormal distribution.
from scipy.stats import lognorm
fig, ax = plt.subplots(1, 1)
s = 0.5
x = np.linspace(lognorm.ppf(0.01, s),
                lognorm.ppf(0.99, s), 100000)
ax.plot(x, lognorm.pdf(x, s),
       'r-', lw=5, alpha=0.6, label='lognorm pdf')
plt.xlabel('$x$')
plt.ylabel('$lognormal cdf(x) with s=0.5$')
# Here is a lognormal graph with s = 0.5
# s will determine the steepness of the peak
# and the position of the peak.
```

```python
# 2. B. cont.
fig, ax = plt.subplots(1, 1)
s = 0.8
x = np.linspace(lognorm.ppf(0.01, s),
                lognorm.ppf(0.99, s), 100000)
ax.plot(x, lognorm.pdf(x, s),
       'r-', lw=5, alpha=0.6, label='lognorm pdf')
plt.xlabel('$x$')
plt.ylabel('$lognormal cdf(x) with s=0.8$')
# Here is a lognormal graph with s = 0.8
# when s increases, the the peak gets more 
# steep but lower and is shifted leftward.
```

```python
# 2. B. cont.
fig, ax = plt.subplots(1, 1)
s = 0.2
x = np.linspace(lognorm.ppf(0.01, s),
                lognorm.ppf(0.99, s), 100000)
ax.plot(x, lognorm.pdf(x, s),
       'r-', lw=5, alpha=0.6, label='lognorm pdf')
plt.xlabel('$x$')
plt.ylabel('$lognormal cdf(x) with s=0.2$')
# Here is a lognormal graph with s = 0.2
# when s decreases, the the peak gets less 
# steep but higher and is shifted rightward.
```
