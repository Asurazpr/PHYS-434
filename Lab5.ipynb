```python
def normalize(data):
    result = np.array([])
    minVal = min(data)
    maxVal = max(data)
    result = (data-minVal)/(maxVal - minVal)
    return result
bkgd_mass = normalize(bkgd_dict1['mass'])
sig_mass = normalize(sig_dict1['mass'])
```

```python
x = np.linspace(0, 100000, 100000)
fig, ax = plt.subplots(1, 1)
ax.plot(x, bkgd_mass, 'b-', lw=1, alpha = 0.3)
ax.plot(x, sig_mass, 'r-', lw=1, alpha = 0.3)
plt.show()
```
