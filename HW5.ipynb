```python
import numpy as np
from scipy import stats
from scipy.stats import norm
from scipy.stats import uniform
import matplotlib.pyplot as plt
```

```python
bkgd = norm.rvs(scale = 2, size=1000000)
signal = uniform.rvs(0, 20, size=1000000)
```

```python
data = bkgd + signal

signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-7,27,68)

Psd, temp, temp2= np.histogram2d(data,signal, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
plt.show()
```

```python
plt.hist(data[sigidx], bins = 45, edgecolor = 'black', linewidth = 1.2) 
plt.yscale('log')                                                       
plt.title('True Signal = 5', size = 22)
plt.xlabel('Observed Signal', size = 20)
plt.show()
```

We pick the true signal to be 5 and the resulting probabilities of observed signal are shown in the above plot. From above graph, we can see that the log plot of the signal looks like a upside-down parabola which indicates a Gaussian distribution.
