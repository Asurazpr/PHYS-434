## Problem 0

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.stats import norm
```

```python
t = np.linspace(0, np.pi*100, int(1e5))
N = 10 + np.sin(t)
bkgd = norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

## Problem 1: subsets of data
#### 1a)

```python
fig,ax = plt.subplots(1, 1)
ax.plot(t[:1000], bkgd[:1000], 'bo', label='scatter plot of the first 1000 points')
plt.show()
```

#### 1b)

```python
def rangePlot(x, y, start, end):
    fig, ax = plt.subplots(1, 1)
    ax.plot(x[start:end], y[start:end], 'bo', label='scatter plot of the first 1000 points')
    plt.show()
```

```python
rangePlot(t, bkgd, 500, 2000)
```

#### 1c)

```python
def samplePlot(x, y, size):
    fig, ax = plt.subplots(1, 1)
    ax.plot(x[0:int(len(x)):size], y[0:int(len(y)):size], 'bo', label='scatter plot of the first 1000 points')
    plt.show()
```

```python
samplePlot(t, bkgd, 100)
```

## Problem 2: density
#### 2a)

```python
import matplotlib as mpl
```

```python
fig, ax = plt.subplots(figsize=(6, 1))
fig.subplots_adjust(bottom=0.5)

cmap = mpl.cm.viridis
bounds = [-1, 2, 5, 7, 12, 15]
normal = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='both')

fig.colorbar(mpl.cm.ScalarMappable(norm=normal, cmap=cmap),
             cax=ax, orientation='horizontal',
             label="Discrete intervals with extend='both' keyword")
```

```python
fig, ax = plt.subplots(1, 1)
plt.hist2d(t, bkgd, bins=30, cmap=cmap)
plt.show()
```

#### 2b)

The plot shows that as we goes toward the horizontal midline, the density of points increases. From the above colorbar setting, the lighter the color, the denser the points. We can see that there are some places in the plot where the colors blend with each other which demonstrates the change in density in certain areas.
